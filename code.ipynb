```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('coco.png',flags = cv2.IMREAD_UNCHANGED)
```

```python
img.shape
```

```python
plt.figure(figsize=(20,10))
plt.imshow(img)
x_lines = [0,100,195,290,385,480,580,670,770,865,961]
y_lines = [0,110,220,320,420,525,625,725,832]
for x in x_lines:
    plt.axvline(x)
for y in y_lines:
    plt.axhline(y)
# plt.savefig('test.png')
```

```python
import numpy as np
```

```python
stt = 0
for i in range(10):
    for j in range(10):
        x_min = x_lines[i]
        x_max = x_lines[i+1]
        y_min = y_lines[j]
        y_max = y_lines[j+1]
        img_cat = img[x_min:x_max,y_min:y_max,:]
        if img_cat[:,:,3].any():
            cv2.imwrite(f'coco_img/coca_stt_{stt}.png',img_cat)
            stt += 1
```
